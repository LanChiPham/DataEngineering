```python
import findspark
findspark.init()

import pyspark
from pyspark import SparkConf
from pyspark import SparkContext
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.master('local[*]').getOrCreate()
```

```python
# countByValue example
```

```python
sample_text = "singapore is a safe country singapore passport is very powerful singapore is a small country"
```

```python
rdd_base = spark.sparkContext.parallelize(sample_text.split(" "))
```

```python
rdd_base.countByValue()
```
